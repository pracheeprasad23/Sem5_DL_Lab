Implementing Feedforward neural networks in Python using Keras and TensorFlow

Problem: Customer Churning

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Step 2: Load dataset
df = pd.read_csv("/content/drive/MyDrive/Sem5_DL_Lab/Customer-Churn-Records.csv")
df.head()
```

```python
# Step 3: Drop irrelevant columns
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
# Step 4: Encode categorical variables
le_geo = LabelEncoder()
le_gender = LabelEncoder()
le_card = LabelEncoder()
```

```python
df['Geography'] = le_geo.fit_transform(df['Geography'])
df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Card Type'] = le_card.fit_transform(df['Card Type'])
```

```python
# Step 5: Define features and target
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
# Step 6: Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 7: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Step 8: Build the Feedforward Neural Network
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Binary classification
])
```

```python
# Step 9: Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Step 10: Train the model
history = model.fit(X_train, y_train,
                    epochs=20,
                    batch_size=32,
                    validation_split=0.2,
                    verbose=1)
```

```python
# Step 11: Evaluate the model
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
